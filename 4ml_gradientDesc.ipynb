```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent parameters
x = 2  # Starting point
learning_rate = 0.1
iterations = 20  # Fewer iterations to better see the steps
```

```python
# To store x values for each step
x_values = [x]
y_values = [function(x)]
```

```python
# Perform Gradient Descent
for _ in range(iterations):
    grad = derivative(x)
    x = x - learning_rate * grad
    x_values.append(x)
    y_values.append(function(x))
```

```python
# Plotting
x_range = np.linspace(-6, 6, 100)  # Range of x values for the function plot
y_range = function(x_range)
plt.figure(figsize=(10, 6))
```

```python
# Plot the function curve
plt.plot(x_range, y_range, label="y = (x + 3)^2", color="blue")

# Plot the gradient descent steps
plt.scatter(x_values, y_values, color="red", label="Gradient Descent Steps", zorder=5)
plt.plot(x_values, y_values, color="red", linestyle="--", alpha=0.6)

# Mark the minimum point
plt.scatter([-3], [function(-3)], color="green", marker="o", label="Local Minimum (x = -3)")

# Labeling the plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent on y = (x + 3)^2")
plt.legend()
plt.grid(True)
plt.show()
```
